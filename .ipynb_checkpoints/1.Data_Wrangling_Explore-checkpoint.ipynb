# Missing Data

Options for handling missing data include:

+ Option 1: Delete instances (rows) that have missing data:  eg: DF.dropna(subset=["column_name"]

+ Option 2: Drop attributes (columns) with a high percentage of missing values: DF.drop("total_bedrooms", axis=1)

+ Option 3: Impute: fill in these missing values: median = housing["total_bedrooms"].median() housing["total_bedrooms"].fillna(median)

+ Optiona 4: Algorithms such as, XGBoost that can handle missing values with no need for imputation.

## Imputing  Numeric Attributes

* When imputing missing data, you should compute the median value on the training set, and use it to fill the missing values in the training set

* Donâ€™t forget to save the median value that you have computed. You will need it later to replace missing values in the test set when you want to evaluate your system, and also once the system goes live to replace missing values in new data.

Scikit-Learn provides a handy class to take care of missing values: Imputer

Computed the median, mode, etc.. of each attribute.  Transform the training set by replacing missing values by the learned medians

```python
# Example u
imputer = Imputer(strategy="median", verbose= True)
imputer.fit(df[num_vars]) 
X = imputer.transform(df[num_vars].copy())
temp = pd.DataFrame(X, columns=num_vars) 
msno.matrix(temp) # sanity check, missing values resplaced ...
```
